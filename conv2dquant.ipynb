```python
import torch
import torchvision
import torchvision.transforms as transforms
from torchvision.models import resnet18

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])
testset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=100, shuffle=False, num_workers=2)

model = resnet18()
model.fc = torch.nn.Linear(model.fc.in_features, 10)
model.load_state_dict(torch.load('resnet18_cifar10.pth'))
model.to(device).eval()

correct, total = 0, 0
with torch.no_grad():
    for images, labels in testloader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f'Accuracy of ResNet18 on CIFAR-10 test images: {100 * correct / total:.2f}%')
```

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np

def uniform_quantization(num_bits):
    def quant(x: torch.Tensor):
        min_val = x.min()
        max_val = x.max()
        alpha = max_val - min_val
        x = (x - min_val) / alpha
        scale = (2**num_bits - 1)
        result = (scale * x).round()
        result /= scale
        result = alpha * result + min_val
        return result
    return quant

def quantize_conv2d(conv_layer: nn.Conv2d, num_bits: int):
    quantize = uniform_quantization(num_bits)
    quantized_weights = quantize(conv_layer.weight.data)
    return quantized_weights

# Create a Conv2d layer
conv = nn.Conv2d(3, 64, kernel_size=(7, 7), stride=(2, 2), padding=(3, 3), bias=False)

# Get original weights
original_weights = conv.weight.data.flatten().numpy()

# Quantize weights with different bit depths
quantized_weights_8bit = quantize_conv2d(conv, 8).flatten().numpy()
quantized_weights_4bit = quantize_conv2d(conv, 4).flatten().numpy()

# Create the plot
fig, axs = plt.subplots(3, 1, figsize=(10, 15))
fig.suptitle('Effect of Quantization on Conv2d Weights')

# Plot original weights
axs[0].hist(original_weights, bins=50, color='blue', alpha=0.7)
axs[0].set_title('Original Weights')
axs[0].set_xlabel('Weight Value')
axs[0].set_ylabel('Frequency')

# Plot 4-bit quantized weights
axs[1].hist(quantized_weights_4bit, bins=50, color='green', alpha=0.7)
axs[1].set_title('4-bit Quantized Weights')
axs[1].set_xlabel('Weight Value')
axs[1].set_ylabel('Frequency')

# Plot 8-bit quantized weights
axs[2].hist(quantized_weights_8bit, bins=50, color='red', alpha=0.7)
axs[2].set_title('8-bit Quantized Weights')
axs[2].set_xlabel('Weight Value')
axs[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
